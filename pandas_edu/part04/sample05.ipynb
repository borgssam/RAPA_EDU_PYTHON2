```python
import seaborn as sns
 
# auto mpg 데이터셋 로드
auto_mpg_data = sns.load_dataset('mpg')
 
# 연비, 실린더 수, 중량만 선택
selected_columns = auto_mpg_data[['mpg', 'cylinders']]
 
# 기술 통계 계산
mean_values = selected_columns.mean() # 평균값
median_values = selected_columns.median() # 중간값
max_values = selected_columns.max() # 최대값
min_values = selected_columns.min() # 최소값
std_dev = selected_columns.std() # 표준편차
correlation = selected_columns.corr()
 
selected_columns.corr() # 상관계수
# 결과 출력
print("평균값:\n", mean_values)
print("\n중간값:\n", median_values)
print("\n최대값:\n", max_values)
print("\n최소값:\n", min_values)
print("\n표준편차:\n", std_dev)
print("\n상관계수:\n", correlation)
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
# 데이터셋 로드
#auto_mpg_data = sns.load_dataset('mpg')
# 데이터 정렬
sorted_data = auto_mpg_data.sort_values(by='weight')
# 선 그래프 그리기
plt.plot(sorted_data['weight'], sorted_data['mpg'], color='blue', 
marker='o', linestyle='-', markersize=4)
plt.title('Weight vs MPG')
plt.xlabel('Weight')
plt.ylabel('MPG')
plt.grid(True)
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
 
# auto mpg 데이터셋 로드
auto_mpg_data = sns.load_dataset('mpg')
 
# 데이터 정렬
sorted_data = auto_mpg_data.sort_values(by='weight')
 
# 면적 그래프 그리기
plt.fill_between(sorted_data['weight'], sorted_data['mpg'], color='skyblue', alpha=0.5)
# plt.plot(sorted_data['weight'], sorted_data['mpg'], color='blue', linewidth=2)
plt.title('Weight vs MPG (Area Plot)')
plt.xlabel('Weight')
plt.ylabel('MPG')
plt.grid(True)
plt.show()
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
# auto mpg 데이터셋 로드
# auto_mpg_data = sns.load_dataset('mpg')
# 실린더 수별 평균 연비 계산
cyl_mpg = auto_mpg_data.groupby('cylinders')['mpg'].mean()
# 막대 그래프 그리기
cyl_mpg.plot(kind='barh', color='lightgreen', edgecolor='black')
plt.title('Average MPG by Cylinders')
plt.xlabel('Cylinders')
plt.ylabel('Average MPG')
plt.xticks(rotation=0)
plt.show()
```
