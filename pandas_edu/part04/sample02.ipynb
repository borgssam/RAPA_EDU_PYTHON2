```python
import pandas as pd 
import matplotlib.pyplot as plt 

samsung_revenue = pd.read_csv("./data/삼성전자_분기별_매출액.csv")
samsung_revenue = samsung_revenue.sort_values('quarter')
samsung_revenue
```

```python
fig = plt.figure()
# print(fig)
fig, axe = plt.subplots(figsize=(8,2))
axe.plot(samsung_revenue['quarter'],samsung_revenue['value'])
plt.show()
# print(fig)
# print(axe)
```

```python
fig = plt.figure()
# print(fig)
fig, axes = plt.subplots(1,2, figsize=(12,2))
axes[0].plot(samsung_revenue['quarter'],samsung_revenue['value'])
# axes[1].plot(samsung_revenue['value'],samsung_revenue['quarter'])
samsung_revenue['value'].plot(ax=axes[1])

# axes[0].
plt.show()
# print(fig)
# print(axe)
```

```python
fig = plt.figure()
fig, axes = plt.subplots(2,2)
axes[1,1].plot(samsung_revenue['quarter'],samsung_revenue['value'])
plt.show()
```

```python
fig, axes = plt.subplot_mosaic([['top_left', 'right'],
                               ['bottom_left', 'right']])
axes['bottom_left'].plot(samsung_revenue['quarter'], samsung_revenue['value'])
plt.show()
```
