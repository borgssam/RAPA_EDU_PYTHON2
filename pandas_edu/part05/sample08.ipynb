```python
import pandas as pd 
import matplotlib.pyplot as plt 

data_range = pd.date_range(start='2024-10-01', periods = 10, freq='D')
data = [ 1,2,5,8,4,6,4,8,7,8]
df = pd.DataFrame(data, index=data_range, columns=["value"] )
df.plot(kind='line' , title='sample chart')
plt.xlabel('date')
plt.ylabel('value')
plt.show()
```
